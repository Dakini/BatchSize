```python
from fastai import *
from fastai.vision import *
```

```python
path = untar_data(URLs.IMAGENETTE_160)
```

```python
bs = 512
size=128
```

```python
np.random.seed(42)
src = ImageList.from_folder(path).split_by_folder(valid='val').label_from_folder()
tfms = get_transforms()
```

```python
data = src.transform(tfms, size=size).databunch(bs=bs).normalize(imagenet_stats)
```

```python
data.show_batch()
```

```python
data = src.transform(tfms, size=size).databunch(bs=bs).normalize(imagenet_stats)
learn = cnn_learner(data, models.resnet34, metrics=accuracy, callback_fns=ShowGraph)
learn.fit_one_cycle(5)
```
