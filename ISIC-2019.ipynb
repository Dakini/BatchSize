```python
from fastai import *
from fastai.vision import *
```

```python
path = Path('/home/david/Documents/ISIC/')
```

```python
bs = 512
size=128
```

```python
np.random.seed(42)
src = (ImageList.from_csv(path, 'training.csv', folder='train', suffix='.jpg')
       .split_by_rand_pct(0.2)
       .label_from_df())
tfms = get_transforms()
```

```python
data = src.transform(tfms, size=size).databunch(bs=bs).normalize(imagenet_stats)
```

```python
data.show_batch()
```

```python
data = src.transform(tfms, size=size).databunch(bs=bs).normalize(imagenet_stats)
learn = cnn_learner(data, models.resnet34, metrics=accuracy, callback_fns=ShowGraph)
learn.fit_one_cycle(5)
```
