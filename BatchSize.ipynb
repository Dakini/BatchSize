```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from fastai import *
from fastai.vision import *
from fastai.callbacks.hooks import *
from fastai.callbacks import *
import os
```

```python
path = Path('../animeface-character-dataset')
```

```python
gc.collect()
```

```python
bs = 728
size=128
```

```python
np.random.seed(42)
src = ImageList.from_folder(path/'thumb').split_by_rand_pct(0.2).label_from_folder()
tfms = get_transforms()
```

```python
data = src.transform(tfms, size=size).databunch(bs=bs).normalize(imagenet_stats)
```

```python
data.show_batch()
```

```python
learn = cnn_learner(data, models.resnet34, metrics=accuracy, callback_fns=ShowGraph)
learn.fit_one_cycle(5)
```
